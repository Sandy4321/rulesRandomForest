```python
from randomForest import RandomForest
import pandas as pd
```

```python
df = pd.read_csv("data/titanic.csv", sep="\t")
df = df.fillna("nan")
cols = ["Age", "Sex", "Embarked", "Fare", "Pclass", "Survived"]
df = df[cols]
df.head()
```

```python
randomForest = RandomForest()
randomForest.load_pandas(df)
randomForest.fit(antecedents = ["Age", "Sex", "Embarked", "Fare", "Pclass"], consequent = "Survived", desired_state = 1)
frame = randomForest.get_frame()
```
